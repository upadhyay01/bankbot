```python
import os

# Specify the path to your training and testing files
base_path_training = "C:/Users/prave/OneDrive/Desktop/n_p/Training"
base_path_testing = "C:/Users/prave/OneDrive/Desktop/n_p/Testing"

# List of dance styles
dance_styles = ["American Waltz", "Bachata", "Cha Cha", "Lindy Hop", "Merengue", "Salsa"]

# Function to display images in a dance style folder
def display_images(base_path, dance_style):
    style_path = os.path.join(base_path, dance_style)
    
    # Check if the folder exists
    if os.path.exists(style_path):
        print(f"Images in {dance_style} ({os.path.basename(base_path)}):")
        for filename in os.listdir(style_path):
            if filename.endswith(".png"):
                print(os.path.join(style_path, filename))
    else:
        print(f"{dance_style} folder not found in {os.path.basename(base_path)}.")

# Display images for each dance style in training data
for style in dance_styles:
    display_images(base_path_training, style)

# Display images for each dance style in testing data
for style in dance_styles:
    display_images(base_path_testing, style)
```

```python
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Specify the path to your training and testing data
base_path_training = "C:/Users/prave/OneDrive/Desktop/n_p/Training"
base_path_testing = "C:/Users/prave/OneDrive/Desktop/n_p/Testing"

# Image dimensions and batch size
img_width, img_height = 64, 64
batch_size = 32

# Creating ImageDataGenerator for data augmentation
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)

# Generating training data
train_generator = train_datagen.flow_from_directory(
    base_path_training,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical'
)

# Generating testing data
test_generator = test_datagen.flow_from_directory(
    base_path_testing,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical'
)

# Building the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(img_width, img_height, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(len(dance_styles), activation='softmax'))  # Output layer with the number of dance styles

# Compiling the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Training the model and storing the history
history = model.fit(train_generator, epochs=10, validation_data=test_generator)

# Evaluate the model on the testing data
accuracy = model.evaluate(test_generator)[1]
print(f"Accuracy on testing data: {accuracy * 100:.2f}%")

# Plotting training accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Training', 'Validation'], loc='upper left')
plt.show()

# Save the model for future use
model.save("dance_classification_model.h5")
```
